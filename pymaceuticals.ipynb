```python
# dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# load Mouse and Drug data into a data frame
mouse_drug_df = pd.read_csv('data/mouse_drug_data.csv')
mouse_drug_df.head()
```

```python
# load Clinical Trial data into a data frame
clinical_trial_df = pd.read_csv('data/clinicaltrial_data.csv')
clinical_trial_df.head()
```

```python
# combine data into a single dataset
merge_df = pd.merge(mouse_drug_df, clinical_trial_df, on='Mouse ID', how='left')
merge_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint into a data frame
mean_tumor_volume_df = merge_df.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean().to_frame()
mean_tumor_volume_df.reset_index(inplace=True)
mean_tumor_volume_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint into a data frame
stderr_tumor_volume_df = merge_df.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem().to_frame()
stderr_tumor_volume_df.reset_index(inplace=True)
stderr_tumor_volume_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tumor_volume_pivot_df = pd.pivot_table(mean_tumor_volume_df,
                                            values='Tumor Volume (mm3)',
                                            index='Timepoint',
                                            columns='Drug',
                                            aggfunc=np.sum)
stderr_tumor_volume_pivot_df = pd.pivot_table(stderr_tumor_volume_df,
                                              values='Tumor Volume (mm3)',
                                              index='Timepoint',
                                              columns='Drug',
                                              aggfunc=np.sum)
mean_tumor_volume_pivot_df
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(x=mean_tumor_volume_pivot_df['Capomulin'].index,
             y=mean_tumor_volume_pivot_df['Capomulin'],
             yerr=stderr_tumor_volume_pivot_df['Capomulin'],
             fmt='r-o')
plt.errorbar(x=mean_tumor_volume_pivot_df['Infubinol'].index,
             y=mean_tumor_volume_pivot_df['Infubinol'],
             yerr=stderr_tumor_volume_pivot_df['Infubinol'],
             fmt='b-^')             
plt.errorbar(x=mean_tumor_volume_pivot_df['Ketapril'].index,
             y=mean_tumor_volume_pivot_df['Ketapril'],
             yerr=stderr_tumor_volume_pivot_df['Ketapril'],
             fmt='g-s')             
plt.errorbar(x=mean_tumor_volume_pivot_df['Placebo'].index,
             y=mean_tumor_volume_pivot_df['Placebo'],
             yerr=stderr_tumor_volume_pivot_df['Placebo'],
             fmt='k-d')
plt.xticks(range(0, 50, 5))
plt.grid()
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.legend(loc='upper left')

# Save the Figure
plt.savefig("Images/pymaceuticals_tumor_response_to_treatment.png")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean of Metastatic Sites into a data frame. Site data grouped by Drug and Timepoint.
mean_metastatic_sites_df = merge_df.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].mean().to_frame()
mean_metastatic_sites_df.head()
```

```python
# Store the Standard Error associated with Metastatic Sites into a data frame grouped by Drug and Timepoint 
stderr_metastatic_sites_df = merge_df.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem().to_frame()
stderr_metastatic_sites_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_metastatic_sites_pivot_df = pd.pivot_table(mean_metastatic_sites_df,
                                                values='Metastatic Sites',
                                                index='Timepoint',
                                                columns='Drug',
                                                aggfunc=np.sum)
stderr_metastatic_sites_pivot_df = pd.pivot_table(stderr_metastatic_sites_df,
                                                  values='Metastatic Sites',
                                                  index='Timepoint',
                                                  columns='Drug',
                                                  aggfunc=np.sum)
mean_metastatic_sites_pivot_df
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(x=mean_metastatic_sites_pivot_df['Capomulin'].index,
             y=mean_metastatic_sites_pivot_df['Capomulin'],
             yerr=stderr_metastatic_sites_pivot_df['Capomulin'],
             fmt='r-o')
plt.errorbar(x=mean_metastatic_sites_pivot_df['Infubinol'].index,
             y=mean_metastatic_sites_pivot_df['Infubinol'],
             yerr=stderr_metastatic_sites_pivot_df['Infubinol'],
             fmt='b-^')             
plt.errorbar(x=mean_metastatic_sites_pivot_df['Ketapril'].index,
             y=mean_metastatic_sites_pivot_df['Ketapril'],
             yerr=stderr_metastatic_sites_pivot_df['Ketapril'],
             fmt='g-s')             
plt.errorbar(x=mean_metastatic_sites_pivot_df['Placebo'].index,
             y=mean_metastatic_sites_pivot_df['Placebo'],
             yerr=stderr_metastatic_sites_pivot_df['Placebo'],
             fmt='k-d')
plt.xticks(range(0, 50, 5))
plt.grid()
plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration (Days)')
plt.ylabel('Metastatic Sites')
plt.legend(loc='upper left')

# Save the Figure
plt.savefig("Images/pymaceuticals_metastatic_response_to_treatment.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric) into a data frame
mice_count_df = merge_df.groupby(['Drug', 'Timepoint'])['Mouse ID'].count().to_frame()
mice_count_df.rename(columns={'Mouse ID': 'Mouse Count'}, inplace=True)
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count_pivot_df = pd.pivot_table(mice_count_df,
                                     values='Mouse Count',
                                     index='Timepoint',
                                     columns='Drug',
                                     aggfunc=np.sum)
mice_count_pivot_df.head()
```

```python
# Generate the Plot (Accounting for percentages)
mice_count_pivot_df[['Capomulin','Infubinol','Ketapril','Placebo']].plot(kind='line',
                   style={'Capomulin': 'r-o',
                          'Infubinol': 'b-^',
                          'Ketapril':  'g-s',
                          'Placebo':   'k-d'},
                   xlim=(-2.5, 47.5),
                   title='Survival During Treatment',
                   grid=True)
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rate (%)')
plt.legend(title='')

# Save the Figure
plt.savefig("Images/pymaceuticals_survival_rate.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_changes = 100*(mean_tumor_volume_pivot_df.iloc[9]-mean_tumor_volume_pivot_df.iloc[0])/mean_tumor_volume_pivot_df.iloc[0]
percent_changes
```

```python
# Store all Relevant Percent Changes into a Tuple
percent_changes_to_keep = percent_changes.loc[['Capomulin','Infubinol','Ketapril','Placebo']]

# Splice the data between passing and failing drugs
passing_drugs = percent_changes_to_keep[percent_changes_to_keep<0]
failing_drugs = percent_changes_to_keep[percent_changes_to_keep>=0]

# Orient widths. Add labels, tick marks, etc.
fig, ax = plt.subplots()

# Create bars for successful drugs
ax.bar(passing_drugs.index, passing_drugs, color='green', align='edge', width=-1)
x_pos = np.arange(-1*len(passing_drugs), 0, 1) + 0.5
position_percent_tuples = list(zip(x_pos, passing_drugs))
for pos, label in position_percent_tuples:
    plt.text(pos, -3, '{0:.0f}%'.format(label), color='white', ha='center', va='top')

# Create bars for failing drugs
ax.bar(failing_drugs.index, failing_drugs, color='red', align='edge', width=-1)
x_pos = np.arange(0, len(failing_drugs)) + 0.5
position_percent_tuples = list(zip(x_pos, failing_drugs))
for pos, label in position_percent_tuples:
    plt.text(pos, 3, '{0:.0f}%'.format(label), color='white', ha='center', va='bottom')

ax.set(title='Tumor Change Over 45 Day Treatment',
       ylim=[-25, 65],
       xlabel='Drug',
       ylabel='% Tumor Volume Change')
labels = ax.get_xticklabels()
plt.setp(labels, horizontalalignment='right')
ax.grid(True)


# Save the Figure
plt.savefig("Images/pymaceuticals_tumor_change.png")

# Show the Figure
plt.show()
```
